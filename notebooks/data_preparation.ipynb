# 3.Colonnes constantes et quasi-constantes
## 3.1.Analyse des colonnes constantes et quasi-constantes

```python
# vérification des colonnes constantes
constant_features = [col for col in train_df.columns if train_df[col].nunique(dropna=False) <= 1]

print("Nombre de colonnes constantes :", len(constant_features))
constant_features[:30]
```

```python
# Détection des colonnes quasi-constantes
quasi_constant_features = []

for col in train_df.columns:
    
    # fréquence de la modalité la plus présente
    predominant = train_df[col].value_counts(normalize=True, dropna=False).values[0]
    
    # seuil à 99%
    if predominant > 0.99:
        quasi_constant_features.append(col)

print("Nombre de colonnes quasi-constantes :", len(quasi_constant_features))
quasi_constant_features[:20]
```

## 3.2.Suppression des colonnes constantes et quasi-constantes

```python
# suppression des colonnes constantes
train_df = train_df.drop(columns=quasi_constant_features)

print("Nouvelle shape :", train_df.shape)
```

# 4.Colonnes corrélées entre elles

```python
# Colonnes numériques
numeric_df = train_df.select_dtypes(include=["number"])
```

```python
# matrice de corrélation
corr_matrix = numeric_df.corr().abs()
```

```python
# garder le triangle sup
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
# détection des colonnes très corrélées
high_corr_features = [col for col in upper.columns if any(upper[col] > 0.9)]

print("Nombre de colonnes très corrélées :", len(high_corr_features))
high_corr_features[:20]
```

```python
# corrélation avec la target
target_corr = numeric_df.corr()["TARGET"].abs()
```

```python
# seuil de corrélation
threshold = 0.9

# variables à supprimer
drop_cols = set()

# matrice de corrélation absolue
corr_matrix = numeric_df.corr().abs()

# corrélation avec TARGET
target_corr = corr_matrix["TARGET"]

# parcours des colonnes
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        
        col1 = corr_matrix.columns[i]
        col2 = corr_matrix.columns[j]
        
        # corrélation entre deux variables
        if corr_matrix.iloc[i, j] > threshold:
            
            # garder la variable la plus corrélée à TARGET
            if target_corr[col1] < target_corr[col2]:
                drop_cols.add(col1)
            else:
                drop_cols.add(col2)

print("Nombre de colonnes supprimées :", len(drop_cols))
list(drop_cols)[:20]
```

```python
# Suppression des colonnes
train_df = train_df.drop(columns=drop_cols)
```

```python
train_df.shape
```

```python
# heatmap des features les plus corrélées avec TARGET
top_features = target_corr.abs().sort_values(ascending=False).head(15).index.tolist()
top_features = [col for col in top_features if col in train_df.columns]

plt.figure(figsize=(12, 8))
sns.heatmap(train_df[top_features].corr(), annot=True, fmt=".2f", cmap="coolwarm")

plt.title("Corrélation des principales features avec TARGET")
plt.show()
```

```python
# refaire les calculs après nettoyages des colonnes corrélées

# Colonnes numériques
numeric_df = train_df.select_dtypes(include=["number"])

# matrice de corrélation
corr_matrix = numeric_df.corr().abs()

# garder le triangle sup
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# détection des colonnes très corrélées
high_corr_features = [col for col in upper.columns if any(upper[col] > 0.9)]

print("Nombre de colonnes très corrélées :", len(high_corr_features))
high_corr_features[:20]
```

```python
# vérification des types de colonnes restantes
train_df.dtypes.value_counts()
```

```python
# object colonnes
object_cols = train_df.select_dtypes(include=["object"]).columns

print(object_cols.tolist())
```

```python
# Colonnes object restantes
object_cols = train_df.select_dtypes(include=["object"]).columns

# Conversion True/False/NaN en 1/0/NaN
for col in object_cols:
    train_df[col] = train_df[col].map({
        True: 1,
        False: 0
    })
```

```python
# transformation des booleen en int
bool_cols = train_df.select_dtypes(include=["bool"]).columns

train_df[bool_cols] = train_df[bool_cols].astype(int)

train_df.dtypes.value_counts()
```

# 6. Exportation des colonnes catégorielles

```python

from src.data_loading import load_csv

os.makedirs("../data", exist_ok=True)

files = {
    "application_train": "application_train.csv",
    "bureau": "bureau.csv",
    "bureau_balance": "bureau_balance.csv",
    "previous_application": "previous_application.csv",
    "POS_CASH_balance": "POS_CASH_balance.csv",
    "installments_payments": "installments_payments.csv",
    "credit_card_balance": "credit_card_balance.csv",
}

cat_rows = []

for source_name, file_name in files.items():
    df_tmp = load_csv(file_name, nrows=1000)

    cat_cols = df_tmp.select_dtypes(
        include=["object", "category"]
    ).columns.tolist()

    for col in cat_cols:
        cat_rows.append({
            "source_file": source_name,
            "original_categorical_column": col
        })

cat_cols_df = pd.DataFrame(cat_rows)

cat_cols_df.to_csv(
    "../data/processed/original_categorical_columns.csv",
    index=False
)

cat_cols_df
```
